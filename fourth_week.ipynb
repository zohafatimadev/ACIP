**Question no 3:**

Check to see if infact there is any correlation between Calories and other independent variables by
plotting a correlation matrix next

```python
data = df[columns]

plt.figure(figsize=(10,6))
sb.heatmap(data.corr(), annot= True)
plt.show()
```

**Question no 5:**

Figure out which exact item contains a high quantity for ['Calories', 'Total Fat', 'Carbohydrates',
'Dietary Fiber', 'Sugars', 'Protein', 'Vitamin A (% Daily Value)', 'Vitamin C (% Daily Value)', 'Calcium
(% Daily Value)', 'Iron (% Daily Value)'.

```python
for column in columns:
    maximum = df[column].max()
    item_name = df[df[column]==maximum]['Item'].iloc[0]
    print(f'{column}: {item_name}: {maximum}')
```

```python
sb .set(style='whitegrid')
plt.figure(figsize=(30,25))

plt.subplot(5, 2, 1)
sb.stripplot(data=df,x='Category',y='Calories',palette='bright')
plt.xticks(rotation=25,ha='right')
plt.title("Calories Content")

plt.subplot(5, 2, 2)
sb .stripplot(data=df,x='Category',y='Total Fat',palette='bright')
plt.title("Total Fat Content")
plt.xticks(rotation=25,ha='right')

plt.subplot(5, 2, 3)
sb .stripplot(data=df,x='Category',y='Carbohydrates',palette='bright')
plt.title("Carbohydrates Content")
plt.xticks(rotation=20,ha='right')

plt.subplot(5, 2, 4)
sb .stripplot(data=df,x='Category',y='Dietary Fiber',palette='bright')
plt.title("Dietary Fiber Content")
plt.xticks(rotation=25,ha='right')

plt.subplot(5, 2, 5)
sb .stripplot(data=df,x='Category',y='Calories',palette='bright')
plt.title("Calories Content")
plt.xticks(rotation=25,ha='right')

plt.subplot(5, 2, 6)
sb.stripplot(data=df,x='Category',y='Protein',palette='bright')
plt.title("Protein Content")
plt.xticks(rotation=25,ha='right')

plt.subplot(5, 2, 7)
sb.stripplot(data=df,x='Category',y='Vitamin A (% Daily Value)',palette='bright')
plt.title("Vitamin A (% Daily Value) Content")
plt.xticks(rotation=25,ha='right')

plt.subplot(5, 2, 8)
sb.stripplot(data=df,x='Category',y='Vitamin C (% Daily Value)', palette='bright')
plt.title("Vitamin C (% Daily Value) Content")
plt.xticks(rotation=25,ha='right')

plt.subplot(5, 2, 9)
sb.stripplot(data=df,x='Category',y='Calcium (% Daily Value)', palette='bright')
plt.title("Calcium (% Daily Value) Content")
plt.xticks(rotation=25,ha='right')

plt.subplot(5, 2, 10)
sb.stripplot(data=df,x='Category',y='Iron (% Daily Value)', palette='bright')
plt.title("Iron (% Daily Value) Content")
plt.xticks(rotation=25,ha='right')

plt.show()
```
